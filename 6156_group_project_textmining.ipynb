```python
# Remove punctuation from the text column
reviews["Cons"] = reviews["Cons"].str.replace('[^\w\s]', '', regex=True)

# convert all words to lowercase
reviews["Cons"] = reviews["Cons"].str.lower()

# removal of stop_words
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")

stop_JP = stopwords.words("english")

reviews["Cons"] = reviews["Cons"].apply(lambda x: " ".join([word for word in str(x).split() if word not in stop_JP]))
# stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
reviews["Cons"] = reviews["Cons"].apply(lambda x: " ".join([st.stem(word)
                                                         for word in x.split()]))

# further remove custom stopwords, which are problem specific
stop_JP += ["would", "may", "must", "one", "upon", "might", "shall", "could"]
reviews["Cons"] = reviews["Cons"].apply(lambda x: " ".join(x for x in x.split()
                                                         if x not in stop_JP))
```

```python
from wordcloud import WordCloud

comment_words = str(' '.join(reviews["Cons"]).split())

# word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(comment_words)


plt.figure(figsize=(8,8))
plt.clf()
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
# further remove custom stopwords, which are problem specific from the word cloud
stop_JP += ["mani", "way", "make", "want", "say", "might","much"]
reviews["Cons"] = reviews["Cons"].apply(lambda x: " ".join([word for word in str(x).split() if word not in stop_JP]))
#removing words like work, ibm - domain specific words which may appear many times but give no context
stop_JP += ["work", "ibm"]
reviews["Cons"] = reviews["Cons"].apply(lambda x: " ".join([word for word in str(x).split() if word not in stop_JP]))
```

```python
corpus_cons = [doc.split() for doc in reviews["Cons"]]
```

```python
import gensim
from gensim import corpora, models
```

```python
dictionary_cons = corpora.Dictionary(corpus_cons)
```

```python
dictionary_cons.filter_extremes(no_below = 2, no_above = 0.75)
```

```python
# Convert the corpus into Document Term Matrix
DFM_cons = [dictionary_cons.doc2bow(doc) for doc in corpus_cons]
```

```python
n_topics = 10
ldamodel_cons = models.LdaModel(DFM_cons, num_topics=n_topics, id2word=dictionary_cons, passes=40)

# Print topics
for topic_id, topic_words in ldamodel_cons.print_topics(num_topics=n_topics, num_words=10):
    print(f"Topic {topic_id}: {topic_words}")
```
